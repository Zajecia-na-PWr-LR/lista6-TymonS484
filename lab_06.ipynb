```python
import pandas as pd
right_columns = ['gender', 'age', 'race', 'importance_same_race',
       'importance_same_religion',
       'field', 'attractive_important',
       'sincere_important', 'intellicence_important', 'funny_important',
       'ambtition_important', 'shared_interests_important',
       'attractive',
       'sincere', 'intelligence', 'funny', 'ambition', 'sports',
       'tvsports', 'exercise', 'dining', 'museums', 'art', 'hiking', 'gaming',
       'clubbing', 'reading', 'tv', 'theater', 'movies', 'concerts', 'music',
       'shopping', 'yoga', 'match']

df = df[right_columns].copy()
```

```python
profile_features = ['gender', 'age', 'race']
df['user_id'] = df.groupby(profile_features, dropna=False).ngroup()


df = df.dropna(subset=['match'])
df['match'] = df['match'].astype(int)

user_counts = df['user_id'].value_counts()
valid_users = user_counts[user_counts > 5].index
df_valid = df[df['user_id'].isin(valid_users)]

test_indices = df_valid.groupby('user_id').sample(n=5, random_state=42).index
train_indices = df.index.difference(test_indices)

X = df.drop(columns=['match'])
y = df['match']

X_train = X.loc[train_indices].copy()
y_train = y.loc[train_indices].copy()
X_test = X.loc[test_indices].copy()
y_test = y.loc[test_indices].copy()

X_train_id = X_train.pop('user_id')
X_test_id = X_test.pop('user_id')
```

### Zadanie 1.3: Model Bazowy i Wizualizacja Problemu


```python
from sklearn.impute import SimpleImputer
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.metrics import classification_report_imbalanced


X_train_num = X_train.select_dtypes(include=np.number)
X_test_num = X_test.select_dtypes(include=np.number)

imputer_num = SimpleImputer(strategy='median')
X_train_imputed = imputer_num.fit_transform(X_train_num)
X_test_imputed = imputer_num.transform(X_test_num)

print("--- Dummy Prior (Większościowy) ---")
dummy_prior = DummyClassifier(strategy='prior')
dummy_prior.fit(X_train_imputed, y_train)
print(classification_report_imbalanced(y_test, dummy_prior.predict(X_test_imputed)))

print("--- Dummy Uniform (Losowy) ---")
dummy_uniform = DummyClassifier(strategy='uniform', random_state=42)
dummy_uniform.fit(X_train_imputed, y_train)
print(classification_report_imbalanced(y_test, dummy_uniform.predict(X_test_imputed)))

print("--- Logistic Regression ---")
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_imputed, y_train)
y_pred_lr = lr.predict(X_test_imputed)
print(classification_report_imbalanced(y_test, y_pred_lr))
```

### Zadanie 1.4: Techniki Resamplingu


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.metrics import classification_report_imbalanced

idx = np.random.choice(len(X_train_imputed), 1500, replace=False)
X_sample, y_sample = X_train_imputed[idx], y_train.iloc[idx]
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_sample)

plt.figure(figsize=(6, 4))
plt.scatter(X_tsne[y_sample == 0, 0], X_tsne[y_sample == 0, 1], label='0', alpha=0.5, s=10)
plt.scatter(X_tsne[y_sample == 1, 0], X_tsne[y_sample == 1, 1], label='1', alpha=0.5, s=10, c='red')
plt.title("t-SNE (1500 próbek) przed resamplingiem")
plt.legend()
plt.show()

smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train_imputed, y_train)

idx_smote = np.random.choice(len(X_smote), 1500, replace=False)
X_smote_sample, y_smote_sample = X_smote[idx_smote], y_smote.iloc[idx_smote]

tsne_smote = TSNE(n_components=2, random_state=42)
X_tsne_smote = tsne_smote.fit_transform(X_smote_sample)

plt.figure(figsize=(6, 4))
plt.scatter(X_tsne_smote[y_smote_sample == 0, 0], X_tsne_smote[y_smote_sample == 0, 1], label='0', alpha=0.5, s=10)
plt.scatter(X_tsne_smote[y_smote_sample == 1, 0], X_tsne_smote[y_smote_sample == 1, 1], label='1', alpha=0.5, s=10, c='red')
plt.title("t-SNE (1500 próbek) PO samym SMOTE")
plt.legend()
plt.show()

resample_pipe = ImbPipeline([
    ('under', RandomUnderSampler(sampling_strategy=0.5, random_state=42)),
    ('smote', SMOTE(random_state=42)),
    ('model', LogisticRegression(max_iter=1000))
])
resample_pipe.fit(X_train_imputed, y_train)
print("--- Pipeline (UnderSampler + SMOTE) ---")
print(classification_report_imbalanced(y_test, resample_pipe.predict(X_test_imputed)))

sme = SMOTEENN(random_state=42)
X_sme, y_sme = sme.fit_resample(X_train_imputed, y_train)

num_samples_sme = min(1500, len(X_sme))
idx_sme = np.random.choice(len(X_sme), num_samples_sme, replace=False)
X_sme_sample, y_sme_sample = X_sme[idx_sme], y_sme.iloc[idx_sme]

tsne_sme = TSNE(n_components=2, random_state=42)
X_tsne_sme = tsne_sme.fit_transform(X_sme_sample)

plt.figure(figsize=(6, 4))
plt.scatter(X_tsne_sme[y_sme_sample == 0, 0], X_tsne_sme[y_sme_sample == 0, 1], label='0', alpha=0.5, s=10)
plt.scatter(X_tsne_sme[y_sme_sample == 1, 0], X_tsne_sme[y_sme_sample == 1, 1], label='1', alpha=0.5, s=10, c='red')
plt.title("t-SNE PO hybrydowym SMOTE+ENN")
plt.legend()
plt.show()

pipe_sme = ImbPipeline([('model', LogisticRegression(max_iter=1000))])
pipe_sme.fit(X_sme, y_sme)
print("--- SMOTE+ENN ---")
print(classification_report_imbalanced(y_test, pipe_sme.predict(X_test_imputed)))
```

### Zadanie 1.5: Modele Ensemble dla Danych Niezbalansowanych


```python
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import RocCurveDisplay, PrecisionRecallDisplay

print("--- BalancedRandomForestClassifier ---")
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=42)
brf.fit(X_train_imputed, y_train)
print(classification_report_imbalanced(y_test, brf.predict(X_test_imputed)))

print("--- RandomForestClassifier (class_weight='balanced') ---")
rf_bal = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_bal.fit(X_train_imputed, y_train)
print(classification_report_imbalanced(y_test, rf_bal.predict(X_test_imputed)))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
RocCurveDisplay.from_estimator(brf, X_test_imputed, y_test, ax=ax1, name="Balanced RF")
RocCurveDisplay.from_estimator(rf_bal, X_test_imputed, y_test, ax=ax1, name="RF (class_weight)")
ax1.set_title("Krzywa ROC")

PrecisionRecallDisplay.from_estimator(brf, X_test_imputed, y_test, ax=ax2, name="Balanced RF")
PrecisionRecallDisplay.from_estimator(rf_bal, X_test_imputed, y_test, ax=ax2, name="RF (class_weight)")
ax2.set_title("Krzywa Precision-Recall")
plt.show()
```

## CZĘŚĆ 2: Kalibracja Modeli i Macierz Kosztów
### Zadanie 2.1: Ewaluacja w kontekście aplikacji


```python
import pandas as pd

y_prob = brf.predict_proba(X_test_imputed)[:, 1]

results = pd.DataFrame({
    'user_id': X_test_id.values,
    'true_match': y_test.values,
    'prob': y_prob
})
results = results.sort_values(by=['user_id', 'prob'], ascending=[True, False])

def calculate_ranking_metrics(df_user, k=3):
    matches = np.where(df_user['true_match'] == 1)[0]
    mrr = 1.0 / (matches[0] + 1) if len(matches) > 0 else 0.0
    top_k = df_user.head(k)
    prec_at_k = top_k['true_match'].mean() if len(top_k) > 0 else 0.0
    return pd.Series({'mrr': mrr, f'prec@{k}': prec_at_k})

metrics_per_user = results.groupby('user_id').apply(calculate_ranking_metrics)
print("Średnie metryki rekomendacyjne dla systemu:")
print(metrics_per_user.mean())
```

### Zadanie 2.2: Strategia "Model Romantyczny"


```python
threshold_romantic = 0.2
y_pred_romantic = (y_prob >= threshold_romantic).astype(int)

from sklearn.metrics import recall_score, precision_score, f1_score
print(f"--- Model Romantyczny (threshold={threshold_romantic}) ---")
print(f"Recall:    {recall_score(y_test, y_pred_romantic):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_romantic):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_pred_romantic):.4f}")
```

### Zadanie 2.3: Strategia "Model Biznesowy"


```python
def calc_business_profit(y_true, y_pred):
    TP = ((y_true == 1) & (y_pred == 1)).sum()
    TN = ((y_true == 0) & (y_pred == 0)).sum()
    FP = ((y_true == 0) & (y_pred == 1)).sum()
    FN = ((y_true == 1) & (y_pred == 0)).sum()
    return (TP * 5) + (FP * -2) + (FN * -1)

thresholds = np.linspace(0, 1, 100)
profits = [calc_business_profit(y_test, (y_prob >= t).astype(int)) for t in thresholds]

best_idx = np.argmax(profits)
threshold_business = thresholds[best_idx]
max_profit = profits[best_idx]

print(f"Najlepszy próg biznesowy: {threshold_business:.2f}")
print(f"Maksymalny zysk biznesowy: {max_profit}")

y_pred_business = (y_prob >= threshold_business).astype(int)
```

### Zadanie 2.4: Podsumowanie Wyników


```python
business_profit_romantic = calc_business_profit(y_test, y_pred_romantic)
business_profit_business = max_profit
business_profit_baseline = calc_business_profit(y_test, brf.predict(X_test_imputed))

summary_data = {
    'Strategia': ['Romantyczna', 'Biznesowa', 'Domyślna (0.5)'],
    'Threshold': [threshold_romantic, threshold_business, 0.5],
    'Recall': [
        recall_score(y_test, y_pred_romantic),
        recall_score(y_test, y_pred_business),
        recall_score(y_test, brf.predict(X_test_imputed))
    ],
    'Precision': [
        precision_score(y_test, y_pred_romantic, zero_division=0),
        precision_score(y_test, y_pred_business, zero_division=0),
        precision_score(y_test, brf.predict(X_test_imputed), zero_division=0)
    ],
    'Zysk Biznesowy': [business_profit_romantic, business_profit_business, business_profit_baseline]
}

display(pd.DataFrame(summary_data))


```

### Wnioski
Model biznesowy ustala odpowiedni threshold, który balansuje liczbę poprawnych
dopasowań w taki sposób, aby odsiać drogie pomyłki (FP) i zoptymalizować realny dochód/zadowolenie.
Model romantyczny zgadza się na stratę finansową za cenę wysokiego odsetka dopasowań.
